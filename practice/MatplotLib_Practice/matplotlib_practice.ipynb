```python
from matplotlib import pyplot as plt
import pandas as pd
import csv
```

```python
ages_x = [i for i in range(25,36)]
dev_y=[38496,42000,46752,49320,53200,56000,62316,64928,67317,68748,73752]
```

```python
py_dev_y =[45372,48876,53850,57287,63016,65998,70003,70000,71496,75370,83640]
js_dev_y=[37810,43515,46823,49293,53437,56373,62375,66674,68745,68746,74583]
```

```python
#print(plt.style.available)#lookup styles
plt.style.use('bmh') # Styles
#plt.xkcd()#another style
plt.plot(ages_x,dev_y,label='All Devs',color='#5a7d9a',linestyle='--', marker='.',linewidth=3 )
plt.plot(ages_x,py_dev_y, label='Python devs', marker='.', color='#444444')
plt.plot(ages_x,js_dev_y, label='Java devs', marker='d', color='#adad3b')

plt.title('Median Salary (USD) by age')
plt.xlabel('ages')
plt.ylabel('Median Salary (USD)')

plt.legend()
plt.grid(True)
plt.tight_layout()#padding imporvement
#plt.savefig('plot.png')#save plot as PNG in current direcotory
```

```python
#BARCHART
import numpy as np
x_indexes=np.arange(len(ages_x))
width = 0.25
plt.bar(x_indexes- width,dev_y,label='All Devs',width=width,color='#5a7d9a', )
plt.bar(x_indexes,py_dev_y, label='Python devs',width=width,color='#444444')
plt.bar(x_indexes + width ,js_dev_y, label='Java devs', width=width, color='#adad3b')
plt.xticks(ticks=x_indexes, labels=ages_x)
plt.title('Median Salary (USD) by age')
plt.xlabel('ages')
plt.ylabel('Median Salary (USD)')
plt.legend()
plt.grid(True)
plt.tight_layout()#padding imporvement
#plt.savefig('plot.png')#save plot as PNG in current direcotory
```
