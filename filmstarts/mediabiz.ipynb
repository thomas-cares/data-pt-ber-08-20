```python
#FUNCTION TURN URL INTO HTML
def make_soup(url):
    html=requests.get(url).content
    return BeautifulSoup(html,'lxml')

url='https://www.mediabiz.de/film/programmplaner/vorschau-filme/0/3?tab=open'
soup_mb=make_soup(url)
soup_mb
```
