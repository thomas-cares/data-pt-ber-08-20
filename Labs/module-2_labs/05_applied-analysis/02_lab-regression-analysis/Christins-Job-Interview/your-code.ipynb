```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
notif=pd.read_csv('logs_notifications.csv')#notifications send
web=pd.read_csv('logs_web.csv')#clicked
```

```python
notif.shape #notifications send
```

```python
notif
```

```python
web.shape #clicked
```

```python
df = pd.merge(notif,web, on = 'ip_anonymized',copy='False', how='left')
df
```

```python
df.rename(columns={'datetime_x':'notification_datetime','browser_x':'notification_browser','language_x':'notification_language','datetime_y':'webvisit_datetime','message_x':'message_type'},inplace = True)
```

```python
df.drop(columns=['browser_y','language_y','message_y'],inplace=True)
```

```python
df.head(9)
```

```python
df['visited']=np.where(df['webvisit_datetime'].isna(),0,1)
```

```python
df.head()
```

```python
dum=pd.get_dummies(df['message_type'])
dum.head()
df_reg=pd.merge(df,dum,left_index=True,right_index=True)

df_reg.rename(columns={0:'message_0',1:'message_1',2:'message_2'},inplace=True)
df_reg
```

```python
X=df_reg[['message_1','message_2']]
Y=df_reg['visited']
X= sm.add_constant(X)

model=sm.Logit(Y,X)
results=model.fit()
results.summary()
#confidence intervals overlap for message 1 and message 2
```

```python
## We want to compare message 1 to message 2 now, since the CI for coefficients above overlap.
X = df_reg[df_reg['message_0'] == 0]['message_2']
Y = df_reg[df_reg['message_0'] == 0]['visited']
X = sm.add_constant(X)
model = sm.Logit(Y,X)
results = model.fit()
results.summary() 
# Message 2 has a negative significant coefficient compared to Message 1.
# We will select Message 1 as the best performing message.
```

```python
###TRANSLATIONS#####
df_reg['count']=1

pivot=pd.pivot_table(data=df_reg, index='notification_language', columns='count',aggfunc='sum' )
pivot['open_rate']=pivot['visited']/pivot['message_type']
pivot
#spanish performing particulary bad we ll use it as base
```
